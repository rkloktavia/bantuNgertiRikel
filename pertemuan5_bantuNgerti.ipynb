## Jenis Matplotlib
### 1. Linechart

```python
### Contoh grafik
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x,y)
plt.show()
```

## 2. Scatter plot

```python
### Scatter plot
x=[1,2,3,4,5]
y=[1,4,124,46,100]
plt.scatter(x,y)
plt.show()
```

### 3. Bar plot

```python
categories=['Nasi campur','sate','bakso','ayam rebus']
valuesHarga=[100,32,45,12]
plt.bar(categories,valuesHarga)
plt.show()
```

nasi campur = 100
sate = 32

### 4. Pie chart / Donut chart

```python
makanan=['nasi','mie','lontong']
pendapatanPenjualan=[12,42,53]
plt.pie(pendapatanPenjualan,labels=makanan)
plt.show()
```

## Parameter Matplolib

### Figsize

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20)) #ukurannya 20px x 20px
plt.plot([1,2,3,4],[6,7,9,100])
plt.show()
```

### DPI (dot per inch)

```python
plt.figure(dpi=100)
plt.plot([1,2,3,4],[6,7,2,1])
plt.show()
```

## PLT title

```python
makanan=['nasi','mie','lontong']
pendapatanPenjualan=[12,42,53]
plt.pie(pendapatanPenjualan,labels=makanan)
plt.title('Pendapatan Penjualan Makanan')
plt.show()
```

## PLT labelling

```python
plt.plot([1,2,3,4],[10,20,25,30])
plt.title('ini adalah Line chart')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

```python
plt.plot([1,2,3,4],[10,23,30,13],color='red')
plt.show()
```

```python
plt.plot([1,2,3,4],[10,23,30,13],linestyle='--')
plt.show()
```

## marker

```python
df.columns
```

```python
### 2. Traffic Source apa yang mendatangkan paling banyak pelanggan ke The Look ECommerce tiap bulannya?
df['traffic_source'].value_counts()
```

```python
### cari month per order
df['order_date']=pd.to_datetime(df['order_date'])
df['month']=df['order_date'].dt.month
print(df['month'].unique())
traffic_per_month=df.groupby(['month','traffic_source'])['id'].nunique().unstack()
print(traffic_per_month)
```

```python
traffic_per_month.plot(kind='line')
plt.xlabel('Bulan')
plt.ylabel('Jumlah traffic source Pelanggan')
plt.title('Pergerakan Traffic source per Bulan')
plt.show()
```

```python
### 1. Dimanakah jumlah pelanggan pria paling banyak, tampilkan dengan bar chart per negara?
print(df.gender.value_counts())
print(df.country.value_counts())
dfPria=df[df['gender']=='M']
print(dfPria.gender.value_counts())
jumlahPriaPerCountry=dfPria.groupby('country')['id'].nunique().sort_values(ascending=False)
print(jumlahPriaPerCountry)
plt.figure(figsize=(10,6))
jumlahPriaPerCountry.plot(kind='bar')
plt.title('Jumlah Pelanggan Pria per Negara')
plt.xlabel('Negara')
plt.ylabel('Jumlah Pelanggan')
plt.show()
```

```python
### 3.Ada berapa total jumlah customer yang ada di dataset ini perbulannya, tampilkan dengan Line Chart?
total_customers=df.groupby('month')['id'].nunique()
```

```python
total_customers
```

```python
plt.figure(figsize=(10,6))
total_customers.plot(kind='line',marker='o')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Pelanggan')
plt.title('Total Pelanggan per Bulan')
plt.show()
```

```python
### 4.Total jumlah order per bulan yang memiliki status order berhasil?
df['month']
```

```python
df.status_order.unique()
```

```python
dfComplete=df[df['status_order']=='Complete']
```
